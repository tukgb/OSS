# 경상남도

```python
gnam = pd.read_csv("상가정보_경남.csv", encoding = "utf-8")

print(gnam)
```

```python
columns = ['상호명', '상권업종대분류명', '상권업종중분류명', '상권업종소분류명', '도로명주소', '구우편번호', '신우편번호', '경도', '위도']
```

```python
gnam_food = gnam[gnam['상권업종대분류명'] == '음식'][columns]
gnam_store = gnam[(gnam['상권업종소분류명'] == '편의점') | (gnam['상권업종소분류명'] == '슈퍼마켓')][columns]
gnam_sports = gnam[gnam['상권업종중분류명'] == '스포츠 서비스'][columns]

gnam_food.to_csv("food_industry_data.csv", mode = "a", index=False, header = False, encoding = "utf-8-sig")
gnam_store.to_csv("convenience_store_data.csv", mode = "a", index=False, header = False, encoding = "utf-8-sig")
gnam_sports.to_csv("sports_center_data.csv", mode = "a", index=False, header = False, encoding = "utf-8-sig")
```
